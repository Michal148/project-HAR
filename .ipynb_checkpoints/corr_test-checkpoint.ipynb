```python
import numpy as np
import pandas as pd
import scipy
import seaborn
import matplotlib.pyplot as plt
```

```python
data_A = pd.read_csv("accelerometerApple.csv")[["seconds_elapsed","x","y","z"]]
#data_A = pd.read_csv("stanie_acc.csv")[["seconds_elapsed","x","y","z"]]
data_A.head()
```

```python
def rock_roll(data):
    N=data.shape[0]
    result_array = np.zeros((N, N))
    for i in range(N):
        result_array[:, i] = np.roll(data[:, 0], i)
    return result_array
```

```python
signal = data_A[["x"]].to_numpy()
temp=signal[11:14]
print(temp)
```

```python
print(rock_roll(temp))
```

```python
t1= rock_roll(temp)
print((np.max(-t1[0,1:])))
```

```python
signal = data_A[["x"]].to_numpy()
w_start, w_max, w_step =  [10, 201, 10]
b_start, b_max, b_step =  [1, round(signal.shape[0]/2), 20]
matrix = np.zeros((int((w_max-w_start)/w_step+1),int((b_max-b_start)/b_step+1)))
for window_length in range(w_start, w_max, w_step):
    for bias in range(b_start, b_max, b_step):
        temp=np.flip(signal[(bias):(bias+window_length)])
        temp1=signal[(-(bias+window_length)):(-bias)]
        x_corr = np.corrcoef(rock_roll(temp),rock_roll(temp1))
        matrix[int((window_length-w_start)/w_step)][int((bias-b_start)/b_step)] =  np.max(np.abs(x_corr[0,1:]))
print(matrix)
```

```python

plt.rcParams["figure.figsize"] = (20,20)
ax = plt.axes()
seaborn.heatmap(np.abs(matrix), ax=ax , yticklabels=np.arange(w_start,w_max,w_step), xticklabels=np.arange(b_start,b_max,b_step))
ax.set_xlabel('bias')
ax.set_ylabel('okno')
plt.show()
```
